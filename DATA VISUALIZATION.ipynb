```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#bar graph
years = [20,30,40,50,60,70,80]
weights = [40,50,60,70,80,90,100]
plt.plot (years,weights, color="brown")
plt.xlabel("years")
plt.ylabel("weight")
```

```python
people =["John","Mary","Sarah","Lisa","James","Gary","Samuel","Daniel","Davis"]
#weights (kgs)
weights = [72,58,56,87,97,56,54,48,65]

#heights (cm)
heights = [175,155,180,160,181,182,165,158,173]

plt.barh(people,weights,color ="green")
plt.title("Weights of students")
plt.xlabel("Weights")
plt.ylabel("Names of students")
```

```python
fig, axes =plt.subplots(1,2, figsize=(12,5))

#scatter plot
axes[0].scatter(people,weights,color= "green", marker="x")
axes[0].set_title("Scatter plot of years vs. Weights")
axes[0].set_xlabel("Year")
axes[0].set_ylabel("Weight (kg)")
axes[0].grid(True)

#bar graph
axes[1].bar(people,heights,color= "green", width=0.5)
axes[1].set_title("Bar chart of years vs. Weights")
axes[1].set_xlabel("Year")
axes[1].set_ylabel("Height (cm)")
axes[1].grid(True)
```

# Seaborn

```python
data = sns.load_dataset("tips")
data
```

```python
sns.histplot(x = "total_bill", y ="tip",hue = "smoker", data=data)
```

```python
sns.histplot(data=data, x= "total_bill",kde=True)
plt.title("Distribution of total bill")
plt.xlabel("Total bill ($)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(data=data, x="total_bill",y= "tip",hue= "smoker",style="sex")
plt.title("Total bill vs. Tip")
plt.xlabel("Total bill ($)")
plt.ylabel("Tip ($)")
plt.legend(title="sex",loc= "upper left")
plt.legend(title="smoker",loc="lower right")
plt.show()
```

```python
sns.barplot(data=data, x="day",y= "tip",errorbar="ci") #ci shows the confidence interval
plt.title("Average Tip by Day")
plt.xlabel("Day of the week")
plt.ylabel("Average Tip($)")
plt.savefig("Average Day Tip.png")
plt.grid(True)
plt.show()
```

```python
sns.boxplot(data=data, x="time",y= "tip",hue="smoker")
plt.title("Tip Amount by Time of Day")
plt.xlabel("Time of Day")
plt.ylabel("Tip($)")
plt.legend(title="smoker")
plt.show()
```

```python
numerical_data = data.select_dtypes(include="number")
numerical_data
```

```python
corr = numerical_data.corr()
corr
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(corr,cmap="coolwarm",cbar=True)
plt.title("Correlation between our numerical_variables")
plt.show()
```
